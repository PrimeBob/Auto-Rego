```python
#opening production sheet
sheet= client.open(f).worksheet("Registration sheet")

df = pd.DataFrame(sheet.get_all_records())
harvestfield=df[df.Registered == 'need6'].reset_index(drop=True)
len(harvestfield)
harvestfield
```

```python
#email column on google sheets
emailloc=2
statusloc=8

##the below code fills in the registration page and submits it
for i in range(len(harvestfield.Email)):        
    print(str(i)+" start check")
    
    try:
        web.refresh()

        #toggle button
        time.sleep(1)
        togbutton=web.find_element_by_xpath('//*[@id="sfid-3116553913"]/a')
        web.execute_script("arguments[0].click();", togbutton)
        time.sleep(2)
            
        time.sleep(3)
                                    
        #first
        first=harvestfield.First[i]
        firstslot=web.find_element_by_xpath('//*[@id="field-1"]')
        
        time.sleep(1)
        firstslot.send_keys(first)
        time.sleep(1)

        #last
        last=harvestfield.Last[i]
        lastslot=web.find_element_by_xpath('//*[@id="field-4"]')
        time.sleep(1)
        lastslot.send_keys(last)
        time.sleep(1)
        
        #email
        email=harvestfield.Email[i]
        emailslot=web.find_element_by_xpath('//*[@id="field-7"]')
        time.sleep(1)
        emailslot.send_keys(email)
        time.sleep(1)  
                
        #company
        organisation=harvestfield.Company[i]
        organisationslot=web.find_element_by_xpath('//*[@id="field-10"]')
        time.sleep(1)
        organisationslot.send_keys(organisation)
        time.sleep(1)

        #title
        title=harvestfield.Title[i]
        titleslot=web.find_element_by_xpath('//*[@id="field-13"]')
        time.sleep(1)
        titleslot.send_keys(title)
        time.sleep(1)
        
        #number
        contactnumber=re.sub("[^0-9]", "",str(harvestfield.Number[i]))
        contactnumberslot=web.find_element_by_xpath('//*[@id="field-16"]')
        time.sleep(1)
        contactnumberslot.send_keys(str(contactnumber))
        time.sleep(1)       

        #cp
        time.sleep(2)
        cprologuebutton=web.find_element_by_xpath('/html/body/div[15]/div/div/div/div/div/div/form/div[2]/div[7]/div/div/div/div[2]/div/div/div[1]/div/div/div[1]')
#                                                   /html/body/div[15]/div/div/div/div/div/div/form/div[2]/div[7]/div/div/div/div[2]/div/div/div[1]/div/div/div[1]
        web.execute_script("arguments[0].click();", cprologuebutton)
        time.sleep(2)

        #country
        country=int(countrydictionary.index(harvestfield.Country[i]))        
        countrybutton=web.find_element_by_xpath('/html/body/div[16]/div/div/div[2]/div/div/div/div['+str(country-1)+']/span')                                                                                                            
        time.sleep(1)
        web.execute_script("arguments[0].click();", countrybutton)
        time.sleep(1)      
        
        #sub button
        time.sleep(2)
        subbutton=web.find_element_by_xpath('/html/body/div[15]/div/div/div/div/div/div/form/div[3]/button/div/div/span/p')
        web.execute_script("arguments[0].click();", subbutton)
        time.sleep(15)
                   
                                        
    except:
        #problem case (when country input has a problem)
        sheet.update_cell(int(df[df.iloc[:,emailloc]==harvestfield.Email[i]].index.to_numpy())+2,statusloc,"problem")
        
        time.sleep(5)
        continue

    #success check    
    try:
        web.find_element_by_xpath('/html/body/div[15]/div/div/div/div/div/div/form/div[3]/p')        
       #web.find_element_by_xpath('//*[@id="sfid-3113699235"]/a').text=='Add to Calendar'or         
        #updating pm sheet        
        sheet.update_cell(int(df[df.iloc[:,emailloc]==harvestfield.Email[i]].index.to_numpy())+2,statusloc,"yes - regoed")        
        time.sleep(4)
        continue 
               
    except:    
        print("")
        #success check    
    if web.current_url=='https://regionalevents.okta.com/oktadigitalidentityforumasia/NowComms':     
        #updating pm sheet        
        sheet.update_cell(int(df[df.iloc[:,emailloc]==harvestfield.Email[i]].index.to_numpy())+2,statusloc,"yes - regoed")        
        time.sleep(4)
        continue 
                
        
    else:
        sheet.update_cell(int(df[df.iloc[:,emailloc]==harvestfield.Email[i]].index.to_numpy())+2,statusloc,"problem")
        time.sleep(4)
        continue   
```
