```python
##############################
#email column on google sheets
##############################
emailloc=7
statusloc=18

##the below code fills in the registration page and submits it
for i in range(len(harvestfield.Email)):        
    print(str(i)+" start check")
    
    try:
        web.delete_all_cookies()
        web.get("https://pages.awscloud.com/execleaders-registration.html?sc_channel=em&sc_campaign=APAC_ENT_WEBINAR_execleaders-21q4_2021103_7014z000000rWeN&sc_medium=em_&sc_content=REG_webinar_field&sc_geo=apac&sc_country=mult&sc_outcome=reg&sc_publisher=other&trk=execleaders21q4_3p_innmedia")       
        
        #tc button
        time.sleep(random.uniform(0,1))
        tcbutton=web.find_element_by_xpath('//*[@id="LblmktoCheckbox_2158792_0"]')
        web.execute_script("arguments[0].click();", tcbutton)

        #sess button
        for j in range(0,4):
            if harvestfield.iloc[i,j]=="YES":
                time.sleep(random.uniform(0,1))
                button=web.find_element_by_xpath('//*[@id="LblmktoCheckbox_2158809_'+str(j)+'"]')
                web.execute_script("arguments[0].click();", button)
                time.sleep(random.uniform(0,1))

        if harvestfield.iloc[i,2]=="YES":
            #session button
            time.sleep(random.uniform(0,1))
            sessionbutton=web.find_element_by_xpath('//*[@id="zOPMarketingFieldMarketoOnly25"]/option[2]')
            sessionbutton.click() 


        if harvestfield.iloc[i,3]=="YES":
            #session button
            time.sleep(random.uniform(0,1))
            sessionbutton=web.find_element_by_xpath('//*[@id="zOPMarketingFieldMarketoOnly26"]/option[2]')
            sessionbutton.click() 


                
        #first
        first=harvestfield.First[i]
        firstslot=web.find_element_by_xpath('//*[@id="FirstName"]')
        time.sleep(random.uniform(0,1))
        firstslot.send_keys(first)
        time.sleep(random.uniform(0,1))

        #last
        last=harvestfield.Last[i]
        lastslot=web.find_element_by_xpath('//*[@id="LastName"]')
        time.sleep(random.uniform(0,1))
        lastslot.send_keys(last)
        time.sleep(random.uniform(0,1))

        #company
        organisation=harvestfield.Company[i]
        organisationslot=web.find_element_by_xpath('//*[@id="Company"]')
        time.sleep(random.uniform(0,1))
        organisationslot.send_keys(organisation)
        time.sleep(random.uniform(0,1))

        #email
        email=harvestfield.Email[i]
        emailslot=web.find_element_by_xpath('//*[@id="Email"]')
        time.sleep(random.uniform(0,1))
        emailslot.send_keys(email)
        time.sleep(random.uniform(0,1))

        #country
        country=int(countrydictionary.index(harvestfield.Country[i]))        
        cbutton=web.find_element_by_xpath('//*[@id="zOPprogressiveprofilingcntry"]/option['+str(country+1)+']')                                                                                                            
        time.sleep(random.uniform(0,1))
        cbutton.click()              
        time.sleep(random.uniform(0,1))

        #state
        if harvestfield.Country[i]=="Australia":
            state=int(statedictionary.index(harvestfield.State[i]))        
            sbutton=web.find_element_by_xpath('//*[@id="zOPprogressiveprofilingsprov"]/option['+str(state+1)+']')                                                                                                            
            time.sleep(random.uniform(0,1))
            sbutton.click()              
            time.sleep(random.uniform(0,1))

        #number
        contactnumber=re.sub("[^0-9]", "",str(harvestfield.Number[i]))
        contactnumberslot=web.find_element_by_xpath('//*[@id="zOPprogressiveprofilingpnum"]')
        time.sleep(random.uniform(0,1))
        contactnumberslot.send_keys(str(contactnumber))
        time.sleep(random.uniform(0,1))

        #comtype
        comtype=int(companytypedictionary.index(harvestfield.CompanyType[i]))        
        combutton=web.find_element_by_xpath('//*[@id="zOPprogressiveprofilingcomptyp"]/option['+str(comtype+1)+']')                                                                                                            
        time.sleep(random.uniform(0,1))
        combutton.click()              
        time.sleep(random.uniform(0,1))

        #city
        city=harvestfield.City[i]
        cityslot=web.find_element_by_xpath('//*[@id="zOPprogressiveprofilingcit"]')
        time.sleep(random.uniform(0,1))
        cityslot.send_keys(city)
        time.sleep(random.uniform(0,1))

        #pc
        pc=harvestfield.Postcode[i]
        pcslot=web.find_element_by_xpath('//*[@id="zOPprogressiveprofilingpcode"]')
        time.sleep(random.uniform(0,1))
        pcslot.send_keys(pc)
        time.sleep(random.uniform(0,1))

        #title
        title=harvestfield.Title[i]
        titleslot=web.find_element_by_xpath('//*[@id="zOPprogressiveprofilingjtitle"]')
        time.sleep(random.uniform(0,1))
        titleslot.send_keys(title)
        time.sleep(random.uniform(0,1))

        #industry
        industry=int(industrydictionary.index(harvestfield.Industry[i]))        
        industrybutton=web.find_element_by_xpath('//*[@id="zOPprogressiveprofilingind"]/option['+str(industry+1)+']')                                                                                                            
        time.sleep(random.uniform(0,1))
        industrybutton.click()              
        time.sleep(random.uniform(0,1))


        try:
            #sub button
            time.sleep(random.uniform(0,1))
            subbutton=web.find_element_by_xpath('//*[@id="mktoForm_50772"]/div[50]/span/button')
            subbutton.click()
            time.sleep(random.uniform(4,7)) 
        
        except:
            pass
        
        try:
            #sub button
            time.sleep(random.uniform(0,1))
            subbutton=web.find_element_by_xpath('//*[@id="mktoForm_50772"]/div[49]/span/button')
            subbutton.click()
            time.sleep(random.uniform(4,7)) 
        
        except:
            pass
        
        try:
            #sub button
            time.sleep(random.uniform(0,1))
            subbutton=web.find_element_by_xpath('//*[@id="mktoForm_50772"]/div[48]/span/button')
            subbutton.click()
            time.sleep(random.uniform(4,7)) 
        
        except:
            pass
        
        try:
            #sub button
            time.sleep(random.uniform(0,1))
            subbutton=web.find_element_by_xpath('//*[@id="mktoForm_50772"]/div[47]/span/button')
            subbutton.click()
            time.sleep(random.uniform(4,7)) 
        
        except:
            pass


    except:
        #problem case (when country input has a problem)
        sheet.update_cell(int(df[df.iloc[:,emailloc]==harvestfield.Email[i]].index.to_numpy())+2,statusloc,"problem")
        print("input problem")
        time.sleep(random.uniform(5,7))
        continue

        

    #success check    
    if web.current_url=="https://pages.awscloud.com/execleaders-registration-ty.html":        
        #updating pm sheet        
        sheet.update_cell(int(df[df.iloc[:,emailloc]==harvestfield.Email[i]].index.to_numpy())+2,statusloc,"yes - regoed")        
        time.sleep(random.uniform(5,7))
        continue 
   
    else:
        sheet.update_cell(int(df[df.iloc[:,emailloc]==harvestfield.Email[i]].index.to_numpy())+2,statusloc,"problem")
        time.sleep(random.uniform(5,7))
        print("success check problem")
        continue   
```
